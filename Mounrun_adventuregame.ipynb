```python
import random
user_name=input("What is your name? ").capitalize()
print(f""" "Welcome to the MONRUN adventure game, {user_name}! \n
You are running away from a monster and you have to choose right direction or compartment or answer the questions correctly
otherwise monster will catch you and you are out of the game.
You can go to either Treasure hunt room or the Lab. \n
In Treasure hunt room you can go either left or right. One direction will make you loose the game while other direction
will lead you to go to room with treasure. This room has 5 numbered written compartments from 1 to 5. one of the compartments
has treasure. You will be given three chances to guess which compartment has treasure. One compartment has monster, you are out
the game if you go to this compartment.\n
In the Lab, you must be careful as there are science experiments taking place. The Lab has two sections,
the Experiment Area and the Storage Area. In each section, you will be asked a question and if you answer
incorrectly then you will be in trouble... \n""")
# creating function for treasure hunt game
shuffle_direction = ["left", "right"]
random.shuffle(shuffle_direction) #suffling the direction
def tresure_hunt():
    counter=0
    direction = input("Which direction do you want to go? type left or right ").lower() #asking user direction
    if direction == shuffle_direction[0]:
        while counter<3:
            counter +=1
            try: ##validating user input
                compartment = int(input(f"Which compartment do you want to go to? choose from 1 to 5 #{counter} chance"))
                if compartment>=5:
                    print("Oh no! monster caught you. You lost the game!")
                    break
                elif compartment==random.randint(1,4):  #randomly generating section which has treasure
                    print("You have found the treasure! You won the game")
                    break
                else:
                    print(f"This is you #{counter} attempt You haven't found the treasure yet.")
                    continue
            except ValueError:
                print("Oops! That was not a valid number. Try again...")
        else:
             print("Sorry you have lost all your life line! You lost the game!")
    elif direction == shuffle_direction[1]:
        print("Sorry you chose the wrong direction! Monster caught you!")
    else:
        print("Please type left or right")
        tresure_hunt()
# creating function for laboratory game
def lab_room():
    lab_section=input("Which section do you want to go? type E for experiment area or S for storage").upper()
    if lab_section == "E":
        question= input("Is hydrogen a gas? type yes or no?").lower()
        if question == "no":
            print("Sorry, you answered incorrectly. Monster got you! lost the game!")
        elif question == "yes":
            print("Congratulations!!!You won! Monster has been thrown away!")
        else:
            print("please type yes or no")
    elif lab_section == "S":
        print("There are some hazardous chemicals in this room")
        ppe_gloves = input("Do you want to wear some gloves and other protective equipments? type yes or no").lower()
        if ppe_gloves == "no":
            print("Oh no the monster has found you... you have lost the game.")
        elif ppe_gloves == "yes":
            print(f"Well done {user_name}, Monster can't catch you. You won!!")
        else:
            print("Please type yes or no")
    else:
            print("Please type E or S only")
            lab_room()
# combining both games together
def user_room_choice():
    user_room = input("which room do you want to go to? type T for Treasure hunt room, L for Laboratory").upper()
    if user_room == "T":
        print("Welcome to the Treasure hunt game. I hope you find the treasure!")
        tresure_hunt()
    elif user_room == "L":
        print("Welcome to the lab! please be careful as you walk around the room.")
        lab_room()
    else:
        print("Please type T or L only")
        user_room_choice()
# main game
while True:
    user_ask = input("Do you want to play this game? type yes or no").lower()
    if user_ask == "yes":
        user_room_choice()
        break
    elif user_ask == "no":
        print("Thank for your time!")
        break
    else:
        print("Please type yes or no")











```

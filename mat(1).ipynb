```python
import json
from datetime import datetime
import matplotlib.pyplot as plt
from collections import Counter

class SafetyHazard:
    def __init__(self, description, risk_level, control_measures):
        self.description = description
        self.risk_level = risk_level
        self.control_measures = control_measures
        self.date_added = datetime.now().strftime("%Y-%m-%d %H:%M")

    def to_dict(self):
        return { "description": self.description,
            "risk_level": self.risk_level,
            "control_measures": self.control_measures,
            "date_added": self.date_added}

class ProcessSafetyManager:
    def __init__(self):
        self.hazards = []

    def add_hazard(self, description, risk_level, control_measures):
        hazard = SafetyHazard(description, risk_level, control_measures)
        self.hazards.append(hazard)
        print(f"[] Hazard added: {description}")

    def list_hazards(self):
        print("\n[] Listing All Safety Hazards:")
        for idx, hazard in enumerate(self.hazards, start=1):
            print(f"\nHazard #{idx}")
            print(f"  Description      : {hazard.description}")
            print(f"  Risk Level       : {hazard.risk_level}")
            print(f"  Control Measures : {hazard.control_measures}")
            print(f"  Date Added       : {hazard.date_added}")

    def export_safety_log(self, filename="safety_log.json"):
        with open(filename, 'w') as file:
            json.dump([h.to_dict() for h in self.hazards], file, indent=4)
        print(f"\n[] Safety log exported to {filename}")

    def visualize_risk_levels(self):
        risk_levels = [hazard.risk_level for hazard in self.hazards]
        risk_count = Counter(risk_levels)

        labels = list(risk_count.keys())
        counts = list(risk_count.values())
        colors = ['green', 'orange', 'red', 'purple']

        plt.figure(figsize=(8, 6))
        plt.bar(labels, counts, color=colors[:len(labels)])
        plt.title("Hazard Risk Level Distribution")
        plt.xlabel("Risk Level")
        plt.ylabel("Number of Hazards")
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

# === Example Usage ===
if __name__ == "__main__":
    manager = ProcessSafetyManager()

    # Add example hazards
    manager.add_hazard("UV exposure from open reactor", "High", "Use shielding and PPE")
    manager.add_hazard("Handling TiO2 nanoparticles", "Medium", "Use fume hood and gloves")
    manager.add_hazard("Electrical short in UV lamp", "Critical", "Install grounding, insulation")
    manager.add_hazard("Spillage of treated water", "Low", "Secondary containment")

    # List and visualize
    manager.list_hazards()
    manager.visualize_risk_levels()
    manager.export_safety_log()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample earthquake magnitude data
# Replace this with your actual DataFrame if loading from a dataset
df = pd.DataFrame({
    'mag': [1.2, 2.8, 3.5, 5.0, 6.8, 7.1, 4.0, 6.2, 3.1, 2.0, 8.3, 4.9, 1.7, 5.6]
})

# Define new custom bins and labels
bins = [0, 2.5, 4.5, 6.5, 10]
labels = ['Very Minor (2.5)', 'Minor (2.54.5)', 'Moderate (4.56.5)', 'Strong (>6.5)']

# Assign categories
df['MagCategory'] = pd.cut(df['mag'], bins=bins, labels=labels, include_lowest=True)

# Count values in each category
mag_counts = df['MagCategory'].value_counts().reindex(labels)

# Create explode effect for the category with the highest count
explode = [0.08 if count == mag_counts.max() else 0 for count in mag_counts]

# Define new, visually appealing color palette
colors = ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728']  # Blue, green, orange, red

# Prepare labels with count + percentage
def make_autopct(values):
    def autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return autopct

# Create pie (donut-style) chart
fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    mag_counts,
    labels=labels,
    autopct=make_autopct(mag_counts),
    startangle=140,
    explode=explode,
    shadow=True,
    colors=colors,
    wedgeprops=dict(width=0.45, edgecolor='w')
)

# Improve readability
plt.setp(autotexts, size=12, weight="bold", color="black")
plt.setp(texts, size=13)

# Add title
plt.title(" Earthquake Magnitude Distribution (Enhanced)", fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as pit
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data setup  use your existing `df` if already loaded
# Assuming df contains 'MagCategory' and 'depth' columns

# Reuse or redefine magnitude categories and labels
bins = [0, 2.5, 4.5, 6.5, 10]
labels = ['Very Minor (2.5)', 'Minor (2.54.5)', 'Moderate (4.56.5)', 'Strong (>6.5)']
df['MagCategory'] = pd.cut(df['mag'], bins=bins, labels=labels, include_lowest=True)

# Set a visual style
sns.set(style="whitegrid")

# Create the figure
plt.figure(figsize=(12, 7))

# Custom palette
palette = {
    'Very Minor (2.5)': '#1f77b4',
    'Minor (2.54.5)': '#2ca02c',
    'Moderate (4.56.5)': '#ff7f0e',
    'Strong (>6.5)': '#d62728'
}

# Create boxplot with Seaborn
sns.boxplot(
    x='MagCategory',
    y='depth',
    data=df,
    palette=palette,
    showmeans=True,
    meanprops={"marker": "o", "markerfacecolor": "black", "markeredgecolor": "black"},
    boxprops=dict(alpha=0.8)
)

# Add plot labels and title
plt.xlabel('Magnitude Category', fontsize=12)
plt.ylabel('Earthquake Depth (km)', fontsize=12)
plt.title(' Depth Distribution by Earthquake Magnitude Category', fontsize=15, weight='bold')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```
